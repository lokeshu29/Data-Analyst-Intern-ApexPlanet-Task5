```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

outliers = df[(df['Sales'] < lower_limit) | (df['Sales'] > upper_limit)]

outliers
```

```python
df = df[(df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'],bins=30)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize =(8,5))
category_sales.plot(kind = 'bar')
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
category_Profit = df.groupby('Category')['Profit'].sum()
plt.figure(figsize=(8,5))
category_Profit.plot(kind = 'bar')
plt.title('Total Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
Sales_trend = df.groupby('Order Date')['Sales'].sum()
plt.figure(figsize =(12,5))
Sales_trend.plot()
plt.title('Total Sales Trend')
plt.xlabel('Order date')
plt.ylabel('Sales')
plt.show()
```

# Key Insights

1. Office Supplies generated the highest total sales among all categories, making it the largest contributor to overall revenue.

2. The Furniture category recorded a negative profit despite having high sales, indicating losses in this category.

3. The sales distribution is right-skewed, where most transactions are of lower values and only a few transactions have very high sales.

4. Discount and Profit have a strong negative correlation (-0.46), suggesting that higher discounts reduce profitability.

5. Sales fluctuate considerably over time, with several peaks and dips, indicating varying customer demand and possible seasonal trends.
