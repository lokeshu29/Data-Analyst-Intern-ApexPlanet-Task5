<b>Business Question 1:</b>
Top 5 Products by Sales

```python
query = """

SELECT

"Product Name",

SUM(Sales) AS TotalSales

FROM superstore

GROUP BY "Product Name"

ORDER BY TotalSales DESC

LIMIT 5;

"""

pd.read_sql(query, conn)
```

<b>Insight</b>

These products contribute the most revenue and can be prioritized for promotions.

<b>Business Question 2:</b>
Monthly Sales Trend

```python
query = """

SELECT

strftime('%Y-%m',"Order Date") AS Month,

SUM(Sales) AS TotalSales

FROM superstore

GROUP BY Month

ORDER BY Month;

"""

pd.read_sql(query, conn)
```

<b>Insight</b>

Sales fluctuate across months, indicating seasonal trends.

<B>Business Question 3:</B>
Customer Segmentation by Spend

Identify which customer segment contributes the highest sales.

<b>Business Question 7:</b>
Average Discount by Category

```python
query = """

SELECT

Category,

AVG(Discount) AS AvgDiscount

FROM superstore

GROUP BY Category;

"""

pd.read_sql(query, conn)
```

<b>Business Question 8:</b>

```python
query = """

SELECT

Region,

COUNT(*) AS TotalOrders

FROM superstore

GROUP BY Region

ORDER BY TotalOrders DESC;

"""

pd.read_sql(query, conn)
```
